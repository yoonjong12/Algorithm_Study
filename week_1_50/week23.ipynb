```python
'''
1. Start
    1-1.   index - w <= 0 :   index + w
    1-2. index - w > 0 :   1 -> count (index-w / (2w+1)) 
2.  index  - start <-> end 
    2-1. start end   ? : (distance / (2w+1))
    2-2.     : continue
3. End 
    3-1.  N   : count (index+w / (2w+1)) 
    3-2.  N  : 
'''
import math

def solution(n, stations, w):
    result = 0
    
    # Start
    start = stations[0] - w
    if start > 1:
        count = math.ceil((start-1) / (2*w+1))
        result += count
        
    # Loop
    for i in range(len(stations) - 1):
        start, end = stations[i]+w, stations[i+1]-w
        if start >= end:
            continue
            
        distance = end-start-1        
        count = math.ceil((distance / (2*w+1)))
        result += count
        
    # End
    end = stations[-1] + w
    if end < n:
        count = math.ceil((n-end) / (2*w+1))
        result += count
        
    return result

n = 11
stations = [4, 11]
w = 1
solution(n, stations, w)
```

```python
'''
1.     pop    
    1-1.  n  m pop 
2.    
3.        
4.     
    4-1. m-1   1  
    
'''
import datetime

def str_to_time(x):
    return datetime.datetime.strptime(x, '%H:%M')

def time_to_str(x):
    hour = '%02d' % x.hour 
    minute = '%02d' % x.minute 
    return hour + ':' + minute

def cut_in_line(n, t, m, timetable):
    last_passenger = timetable[m-1]
    answer = last_passenger - datetime.timedelta(minutes=1)
    answer = time_to_str(answer)
    return answer

def solution(n, t, m, timetable):
    # 22  
    timetable = ['23:59' if i == '24:00' else i for i in timetable]
    
    timetable.sort()
    time =  str_to_time('09:00')
    
    for i in range(n-1):
        count = m
        while count:
            cur_time = str_to_time(timetable[0])
            if cur_time <= time:
                timetable.pop(0)
            count -= 1
                    
        time += datetime.timedelta(minutes=t)
        
    timetable = [str_to_time(i) for i in timetable if str_to_time(i) <= time]
    
    if len(timetable) < m:
        return time_to_str(time)
    
    return cut_in_line(n, t, m, timetable)
    
solution(n, t, m, timetable)
```

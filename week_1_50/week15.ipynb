```python
'''
canJump : N명이 건너려고 할 때 건널 수 있는가? return = T or F
이진탐색으로 건널 수 있는 최대 인원을 구함
'''


def canJump(num, k, stones):
    temp = [i-num if i-num > 0 else 0 for i in stones]
    count = 0
    for i in temp:
        count = 0 if i > 0 else count + 1
        if count == k:
            return False
    return True


def solution(stones, k):
    start = 0
    end = max(stones)
    while start <= end:
        mid = (start + end) // 2
        if canJump(mid, k, stones):
            start = mid + 1
        else:
            end = mid - 1

    return start
```

```python
'''
bfs로 풀자는 생각이 먼저 떠오름
1. words에 begin추가하고, target없으면 0 리턴
2. 각 word를 key로, 1자리만 바꿔서 변환 가능한 word를 리스트를 value로 Dictionary 만듦 (Adjacency Graph)
3. bfs로 target에 처음 도달한 리스트의 길이를 정답으로  반환
'''
def canTransform(X, Y):
    count = 0
    for x, y in zip(X, Y):
        if x != y:
            count += 1
    
    if count == 1:
        return True
    else:
        return False
    
def getAdjGraph(words): # {'abc':['zbc', 'azc', 'abz']} 이런 식
    length = len(words)
    graph = {i:[] for i in words}
    for i in graph.keys():
        for j in words:
            if i != j and canTransform(i, j):
                graph[i].append(j)
    return graph

def graphBFS(graph, begin, target): 
    queue = [(begin, [begin])] 
    while queue: 
        n, path = queue.pop(0) 
        if n == target: 
            return path
        else: 
            for m in set(graph[n]) - set(path): 
                queue.append((m, path + [m]))

def solution(begin, target, words):
    if begin not in words:
        words.append(begin)
    if target not in words:
        return 0 

    graph = getAdjGraph(words)
    result = graphBFS(graph, begin, target)
    
    return len(result) - 1
```

#### 해설

먼저 현재 노드 P의 x값을 Px, 현재 노드의 자식 노드가 가질 수 있는 x 범위를 Lx, Rx (Lx < Px < Rx)라고 하겠습니다. 또 어떤 노드 K의 x값을 Kx 라고 하겠습니다. 만약 현재 노드의 바로 다음 레벨에 Lx ≤ Kx < Px를 만족하는 노드 K가 있다면 K는 노드 P의 왼쪽 자식이 됩니다. 이때, 노드 K의 자식 노드가 가질 수 있는 x값의 범위는 Lx ≤ x ≤ Px – 1 (x ≠ Kx)가 됩니다.

마찬가지로 현재 노드의 바로 다음 레벨에 Px < Kx ≤ Rx를 만족하는 노드 K가 있다면 K는 노드 P의 오른쪽 자식이 되며, 노드 K의 자식 노드가 가질 수 있는 x의 범위는 Px + 1 ≤ x ≤ Rx (x ≠ Kx)가 됩니다.

위 과정을 재귀적으로 반복하면서 각 노드의 왼쪽, 오른쪽 자식을 찾아주면 트리를 구성할 수 있습니다.

노드별 왼쪽, 오른쪽 자식을 찾는 방법은 여러 가지가 있을 수 있습니다. 그중 하나로, 재귀적으로 순회하며 트리를 만들면 같은 level의 노드는 x값이 작은 노드부터 방문하게 되므로, 큐를 트리의 레벨만큼 만들어 두고, x축 기준으로 오름차순 정렬된 노드들을 y축 값이 같은 노드끼리 각 큐에 넣어두면 큐의 front를 확인하는 방법으로 O(1)에 찾을 수 있습니다.

이렇게 하면 노드의 수가 N일 때, 트리를 구성하는 데는 O(N) 시간이 소요되며, 시간 복잡도는 전체 노드를 정렬하는데 걸리는 시간인 O(NlogN)이 됩니다.

```python
'''
전위 순회(preorder) : 루트 -> 왼 -> 오
중위 순회(inorder) : 왼 -> 루트 -> 오
후위 순회(postorder) : 왼 -> 오 -> 루트

1. 리스트를 트리로 만들기 쉽게 정렬 - y축 내림차순, x축 오름차순
2. class로 트리를 빌드

이거 런타임에러로 통과못해서 검색해보니까 setrecursionlimit 이거 설정해야한다카더라
'''
import sys


class Node:
    def __init__(self, num, left=None, right=None):
        self.num = num
        self.left = left
        self.right = right


def getTree(li):
    if len(li):
        root = li[0]
        left = list(filter(lambda x: x[0] < root[0], li))
        right = list(filter(lambda x: x[0] > root[0], li))
        return Node(root[2], getTree(left), getTree(right))
    else:
        return None


def getPreOrder(node):
    global visited
    if not node:
        return
    else:
        visited.append(node.num)
        getPreOrder(node.left)
        getPreOrder(node.right)


def getPostOrder(node):
    global visited
    if not node:
        return
    else:
        getPostOrder(node.left)
        getPostOrder(node.right)
        visited.append(node.num)


def solution(nodeinfo):
    global visited
    sys.setrecursionlimit(1100)

    nodes = [[x[0], x[1], i + 1] for i, x in enumerate(nodeinfo)]
    nodes.sort(key=lambda x: x[0])
    nodes.sort(key=lambda x: x[1], reverse=True)
    root = getTree(nodes)

    answer = []

    # 전위 순회
    visited = []
    getPreOrder(root)
    answer.append(visited)

    # 후위 순회
    visited = []
    getPostOrder(root)
    answer.append(visited)

    return answer
```

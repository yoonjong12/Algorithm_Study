# Add Two Numbers

<div class="content__u3I1 question-content__JfgR"><div><p>You are given two <b>non-empty</b> linked lists representing two non-negative integers. The digits are stored in <b>reverse order</b> and each of their nodes contain a single digit. Add the two numbers and return it as a linked list.</p>

<p>You may assume the two numbers do not contain any leading zero, except the number 0 itself.</p>

<p><b>Example:</b></p>

<pre><b>Input:</b> (2 -&gt; 4 -&gt; 3) + (5 -&gt; 6 -&gt; 4)
<b>Output:</b> 7 -&gt; 0 -&gt; 8
<b>Explanation:</b> 342 + 465 = 807.
</pre>
</div></div>

```python
# Definition for singly-linked list.
# 1. 연결리스트의 모든 값들을 각각 정수로 완성시킴
# 2. 둘이 더함
# 3. 결과를 각 자릿수마다 연결리스트로 만듦

class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
        
l1_1 = ListNode(2)
l1_2 = ListNode(4)
l1_3 = ListNode(3)
l1_1.next = l1_2
l1_2.next = l1_3

l2_1 = ListNode(5)
l2_2 = ListNode(6)
l2_3 = ListNode(4)
l2_1.next = l2_2
l2_2.next = l2_3

class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        val_one = 0
        val_two = 0
        # 자릿수
        i = 1
        
        # 정수 완성
        while l1:
            val_one += l1.val * i
            l1 = l1.next
            i *= 10
            
        i = 1
        while l2:
            val_two += l2.val * i
            l2 = l2.next
            i *= 10
            
        # 둘이 더함
        result = val_one + val_two

        # 결과를 연결리스트로 만듦
        result = str(result)
        
        # 단순 연결 리스트는 헤더 노드가 필요하다
        head = ListNode(int(result[-1]))
        li = head
        for i in range(len(result)-2, -1, -1):
            li.next = ListNode(int(result[i]))
            li = li.next
        
        return head
        

s = Solution()
head = s.addTwoNumbers(l1_1, l2_1)
i = 1
while head:
    print(head.val)
    val_one += head.val * i
    head = head.next
    i *= 10
```

```python
# 1, 깨져야 할 블록들을 1로 교체시킴 이때 1의 개수 만큼 스코어 증가
# 2. 2차원 리스트를 돌리다음 1을 없애고 부족한 만큼 0으로 채워넣음
# 3. 다시 원래대로 돌리면 블록이 빈 공간으로 채워서 내려간 효과!
# 4. 1이 생기지 않을 때까지 반복

def checkBlock(x, y, board):
    left_up = board[x][y]
    left_down = board[x+1][y]
    right_up = board[x][y+1]
    right_down = board[x+1][y+1]
    if left_up+left_down+right_up+right_down == 4 * left_up and left_up != '0':
        return True
    else:
        return False
    
def searchBreak(m, n, board):
    # 탐색
    li = []
    for x in range(m-1):
        for y in range(n-1):
            if checkBlock(x,y,board):
                li.append((x,y))
                
    # 캐쉬에 색칠하기
    for i in li:
        board[i[0]][i[1]] = '1' # left_up
        board[i[0]+1][i[1]] = '1' # left_down
        board[i[0]][i[1]+1] = '1' # right_up
        board[i[0]+1][i[1]+1] = '1' # right_down

    result = 0
    for x in board:
        for y in x:
            if y == '1':
                result += 1
    return result

def solution(m, n, board):
    score = 0
    # 각 단어를 다 분리시킴
    board = [list(i) for i in board]
    
    while True:
        # 깨지는 블록을 찾고 1로 대체
        breaked = searchBreak(m,n,board)
        
        # 더 이상 깨질 것이 없다!
        if breaked == 0:
            break
        else:
            score += breaked

        # 보드를 전치 
        board = list(map(list, zip(*board[::-1])))

        for i in range(n):
            # 1제거 하고 왼쪽으로 다 당겨!
            temp = [item for item in board[i] if item != '1']
            # 부족한 자리는 0으로 채워! 우리가 원하는 길이는 가로행 만큼이니까 m까지
            while len(temp) < m:
                temp += '0'
            board[i] = temp

        board = list(map(list, zip(*board)))[::-1]

    return score
board = ['CCBDE', 'AAADE', 'AAABF', 'CCBBF']
# board = ["TTTANT", "RRFACC", "RRRFCC", "TRRRAA", "TTMMMF", "TMMTTJ"]
m = len(board)
n = len(board[0])
solution(m,n,board) 
```

# Longest Substring Without Repeating Characters

<div><p>Given a string, find the length of the <b>longest substring</b> without repeating characters.</p>

<div>
<p><strong>Example 1:</strong></p>

<pre><strong>Input: </strong><span id="example-input-1-1">"abcabcbb"</span>
<strong>Output: </strong><span id="example-output-1">3 
<strong>Explanation:</strong></span> The answer is <code>"abc"</code>, with the length of 3. 
</pre>

<div>
<p><strong>Example 2:</strong></p>

<pre><strong>Input: </strong><span id="example-input-2-1">"bbbbb"</span>
<strong>Output: </strong><span id="example-output-2">1
</span><span id="example-output-1"><strong>Explanation: </strong>T</span>he answer is <code>"b"</code>, with the length of 1.
</pre>

<div>
<p><strong>Example 3:</strong></p>

<pre><strong>Input: </strong><span id="example-input-3-1">"pwwkew"</span>
<strong>Output: </strong><span id="example-output-3">3
</span><span id="example-output-1"><strong>Explanation: </strong></span>The answer is <code>"wke"</code>, with the length of 3. 
             Note that the answer must be a <b>substring</b>, <code>"pwke"</code> is a <i>subsequence</i> and not a substring.
</pre>
</div>

```python
def sol(s):
    sub_length = 1
    if len(s) == 1:
        return 1
    elif len(s) == 0:
        return 0

    dic = dict()
    index = 0
    start = 0 

    while index < len(s):
        next_token = s[index]
        # 겹치는 단어가 없다
        if not next_token in dic:
            dic[next_token] = index
            index += 1
        else:
            start = dic[next_token] + 1
            index = start
            dic = dict()

        if sub_length < index - start: sub_length = index - start


    return sub_length
sol('abcd')
```

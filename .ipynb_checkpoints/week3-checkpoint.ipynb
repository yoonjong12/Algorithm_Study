## 탑



```python
def solution(h):
    answer = [0 for _ in h]
    for i in range(len(h)-1, 0, -1):
        # 0까지 훑어보려면 -1로 설정
        for j in range(i-1, -1, -1):
            if h[i] < h[j]:
                answer[i] = j+1
                break
    return answer
# reversed = h를 거꾸로 

heights = list(map(int, input()))
answer = solution(heights)
for i in range(len(answer)):
    send_index = [index+1 for index, x in enumerate(answer) if x == i]
    if len(send_index) == 0:
        continue
    elif i == 0:
        print(send_index,'번째 탑이 쏜 신호는 아무도 수신하지 않았습니다.')
    else:
        print(send_index,'번째 탑이 쏜 신호는 '+str(i)+'번째 탑이 수신합니다.')
```

## 스킬트리



```python
# 순서가 지켜진 상태로 문자열을 다 돌면 True
def isSkill(skill, tree):
    # input : 스킬과 스킬트리
    index = 0
    for i in range(len(skill)):
    # enumerate 
        # 스킬의 한 단어가 스킬트리에 존재하는가?
        if skill[i] in tree:
            # 각 스킬단어의 인덱스는 배우는 순서이므로 점점 커져야 한다. c <  b < d
            # 만약 다음 인덱스가 더 작다면 선행스킬이 더 뒤에 있다는 뜻이므로 False
            if index <= tree.index(skill[i]): 
                index = tree.index(skill[i])
            else:
                return False
        # 단어가 존재하지 않다면 뒤의 단어들도 없어야 함 
        elif i+1 < len(skill) and skill[i+1] in tree:
            return False
    return True
# index, element
def solution(skill, skill_trees):
    answer = 0
    for tree in skill_trees:
        if isSkill(skill, tree):
            answer += 1
    return answer
solution('cbd',['bacde','cbadf','aecb','bda'])
```

```python
import sys
input = sys.stdin.readline
 
def bellman_ford():
    for i in range(N):  # 노드 갯수만큼 반복
        for j in range(M):  # 모든간선을 확인하며 거리갱신
            start = edge[j][0]  # 출발
            dest= edge[j][1]  # 도착
            cost = edge[j][2]  # 거리
            cur_cost = dist[start] + cost
            if dist[start] != float('inf') and dist[dest] > cur_cost:  # Relaxation
                dist[dest] = cur_cost
                if i == N-1:  # N일때 변화가 생겼다 -> 음의사이클로인해 무한으로 갱신된다
                    return False
                    
    return True

N, M = map(int,input().split())
 
edge = []
for _ in range(M):
    edge.append(list(map(int, input().split())))  # [출발, 도착, 거리] ...

dist = [float('inf')] * (N+1)
dist[1] = 0

result = bellman_ford()
 
if not result:
    print(-1)
else:
    for i in range(2, N+1):
        cost = dist[i];
        if cost == float('inf'):
            print(-1)
        else:
            print(cost)
```

```python
'''
1. Start
    1-1. 첫째 기지국 index - w <= 0 : 시작 위치가 index + w임
    1-2. index - w > 0 : 시작 위치가 1 -> count에 올림(첫째index-w / (2w+1)) 추가
2. 기지국 index 순회 - start기지국 <-> end기지국 
    2-1. start와 end 사이에 공백이 있는가? : 올림(distance / (2w+1))
    2-2. 두 기지국이 완전히 겹침 : continue
3. End 
    3-1. 기지국이 N까지 덮지 못함 : count에 올림(마지막index+w / (2w+1)) 추가
    3-2. 기지국이 N넘어서 덮어버림 : 끝
'''
import math

def solution(n, stations, w):
    result = 0
    
    # Start
    start = stations[0] - w
    if start > 1:
        count = math.ceil((start-1) / (2*w+1))
        result += count
        
    # Loop
    for i in range(len(stations) - 1):
        start, end = stations[i]+w, stations[i+1]-w
        if start >= end:
            continue
            
        distance = end-start-1        
        count = math.ceil((distance / (2*w+1)))
        result += count
        
    # End
    end = stations[-1] + w
    if end < n:
        count = math.ceil((n-end) / (2*w+1))
        result += count
        
    return result

n = 11
stations = [4, 11]
w = 1
solution(n, stations, w)
```

```python
'''
1. 막차시간 직전까지 승객을 모두 pop해서 막차에 탈 사람들만 남긴다
    1-1. 매 n타임 마다 m명을 pop하는 방식
2. 막차시간 놓친 사람도 제거한다
3. 만약 모두 타도 빈 자리가 남으면 막차시간을 리턴
4. 빈 자리가 없으면 새치기한다 
    4-1. m-1째 승객의 시간에서 1분 일찍 줄선다
    
'''
import datetime

def str_to_time(x):
    return datetime.datetime.strptime(x, '%H:%M')

def time_to_str(x):
    hour = '%02d' % x.hour 
    minute = '%02d' % x.minute 
    return hour + ':' + minute

def cut_in_line(n, t, m, timetable):
    last_passenger = timetable[m-1]
    answer = last_passenger - datetime.timedelta(minutes=1)
    answer = time_to_str(answer)
    return answer

def solution(n, t, m, timetable):
    # 22번 케이스 
    timetable = ['23:59' if i == '24:00' else i for i in timetable]
    
    timetable.sort()
    time =  str_to_time('09:00')
    
    for i in range(n-1):
        count = m
        while count:
            cur_time = str_to_time(timetable[0])
            if cur_time <= time:
                timetable.pop(0)
            count -= 1
                    
        time += datetime.timedelta(minutes=t)
        
    timetable = [str_to_time(i) for i in timetable if str_to_time(i) <= time]
    
    if len(timetable) < m:
        return time_to_str(time)
    
    return cut_in_line(n, t, m, timetable)
    
solution(n, t, m, timetable)
```

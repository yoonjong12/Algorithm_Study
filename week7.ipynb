# 가장 큰 수
<div class="guide-section-description">
      <h6 class="guide-section-title">문제 설명</h6>
      <div class="markdown solarized-dark"><p>0 또는 양의 정수가 주어졌을 때, 정수를 이어 붙여 만들 수 있는 가장 큰 수를 알아내 주세요.</p>

<p>예를 들어, 주어진 정수가 [6, 10, 2]라면 [6102, 6210, 1062, 1026, 2610, 2106]를 만들 수 있고, 이중 가장 큰 수는 6210입니다.</p>

<p>0 또는 양의 정수가 담긴 배열 numbers가 매개변수로 주어질 때, 순서를 재배치하여 만들 수 있는 가장 큰 수를 문자열로 바꾸어 return 하도록 solution 함수를 작성해주세요.</p>

<h5>제한 사항</h5>

<ul>
<li>numbers의 길이는 1 이상 100,000 이하입니다.</li>
<li>numbers의 원소는 0 이상 1,000 이하입니다.</li>
<li>정답이 너무 클 수 있으니 문자열로 바꾸어 return 합니다.</li>
</ul>

```python
# 이 테스트 케이스 덕분에 암이 나았습니다.
numbers = [21,212]

# 1. 문자형 리스트로 변환한다.
# 2. 내림차순으로 숫자의 3배길이로 늘려서 정렬한다.
# 모두 세 글자까지 늘려서 보는 것만 아니라 그 이후까지 고려해야 한다.
def sortFunc(number):
    number *= 3
    return number

def solution(numbers):
    numbers = list(map(str,numbers))
    numbers.sort(key=sortFunc, reverse=True)
    # 11번 테스트케이스 ㅡㅡ
    if all(x == '0' for x in numbers):
        numbers = '0'
    return ''.join(numbers)

solution(numbers)
```

이 경우 어떻게 문자열을 잘라도 압축되지 않으므로 가장 짧은 길이는 17이 됩니다. </p>

```python
# 1. 처음 2부터 문자열의 절반까지 t로 반복
# 2. text는 압축될 문자열
# 3. 조건문1) 단위 길이와 마지막 단위 길이가 다르면 즉, 마지막 부분에서 인덱스를 넘겨서 잘리는 부분은 검사 안함
# 4. token은 검사단어가 들어간 패턴 정규식
# 5. 조건문2) 토큰 너머의 문자열과 매치하는지 검사
# 6. 반복되는 만큼 i를 더해서 검사를 건너뜀
# 7. text에는 건너뛴만큼 숫자를 계산해서 표현을 압축함
# 8. 나머지 케이스들은 정상적으로 토큰을 text에 더함
# 9. 만약 반복문이 s의 길이를 넘기면 종료
# 10. 각 text마다 최소 길이와 비교해서 교체
import re
def pattern(x):
    return re.compile(f'({x})+')

def solution(s):
    min_len = 1000
    # 5번 테스트케이스 ㅡㅡ 더럽
    if len(s)==1:
        return 1
    for t in range(1, len(s)//2+1):
        text = ''
        i = 0
        # 문자열
        while True:
#             print('go',i, s[i])
            # 단위만큼 끊었나?
            if t == len(s[i:i+t]):
                token = pattern(s[i:i+t])
                # 토큰 너머의 문자열과 매치? -> 반복되는가?
                if token.match(s[i+t:]) is not None:
                    end_index = token.match(s[i+t:]).end()
#                     print(token, 'start', i , 'end', end_index)
                    text += f'{(end_index+t)//t}{s[i:i+t]}'
                    i += end_index
#                     print('pass',i)
                else:
                    text += s[i:i+t]
            else:
                text += s[i:i+t]
            i += t
            if i >= len(s):
                break
#         print(text)
        if len(text) < min_len: min_len = len(text)
    return min_len

# solution('aabbaccc')
```

```python
alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
#'NOPQRSTUVWXYZ A BCDEFGHIJKLM'
alphabet = list(alphabet[len(alphabet)//2:] + alphabet[:len(alphabet)//2])

# 1알파벳 최소 탐색 
# 알파벳을 양쪽방향으로 하나씩 인덱스를 옮기면서 탐색
def searchAlphabet(target):
    move = 0
    start = len(alphabet)//2
    while True:
        if alphabet[start-move] == target or alphabet[start+move] == target:
            break
        move += 1
    return move
# cache에서 하나씩 옮기면서 -1이 아닌 즉, 'A'가 아닌 인덱스가 나오면 그 쪽 방향으로 진행
def searchDirection(index,cache):
    step = 0
    while True:
        # right
        if cache[index+step] != -1:
            return True
        # left
        if cache[index-step] != -1:
            return False
        step += 1
        
def solution(name):
    index = 0
    target = 'A' * len(name)
    # 알파벳 검색 여부 캐쉬
    cache = [-1 if i=='A' else 0 for i in name]
    # 알파벳   
    move = 0
    print(name, cache)
    while True:    
        # 알파벳 찾기
        if cache[index] != -1:
            move += searchAlphabet(name[index])
            cache[index] = -1
            
        # 기저 : start랑 타겟이랑 같음?
        print(cache, index, move)
        if set(cache) == {-1}:
            return move
        
        # 방향 찾기
        if searchDirection(index, cache): #left
            index += 1
        else:
            index -= 1
        move += 1
solution('JEROEN')
```
